```python
import numpy as np
import json 
import cv2
import os 
from tqdm import tqdm
```

```python
f = open("mask.json",'r')
data = json.load(f)
data
```

```python
image_dir = 'images'
mask_dir = 'mask'
```

```python
imgs = data['images']
anno = data['annotations']
```

```python
anno
```

```python
for x ,y in zip(imgs,anno):
    file_name = x['file_name']
    h = x['height']
    w = x['width']
    mask = np.zeros((h,w))
    seg = y['segmentation']
    for points in seg:
        contours = []
        #print(points)
        for i in range(0,len(points),2):
            print(i)
            contours.append((points[i],points[i+1]))
        contours = np.array(contours,dtype=np.int32)
        cv2.drawContours(mask,[contours],-1,255,-2)
    cv2.imwrite(f'{mask_dir}/{file_name}',mask)
```
