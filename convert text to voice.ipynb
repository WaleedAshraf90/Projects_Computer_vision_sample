```python
import pydotplus  #function graph 
from IPython.display import display, Image
import matplotlib.pyplot as plt
import pytesseract  #function text 
from gtts import gTTS  #function voice 
import IPython.display as ipd  
from requests import get  # to make GET request
```

```python
#model train arabic:-
def download(url, file_name):
    # open in binary mode
    with open(file_name, "wb") as file:
        # get request
        response = get(url)
        # write to file
        file.write(response.content)
#model 
download("https://github.com/tesseract-ocr/tessdata/raw/master/ara.traineddata","ara.traineddata")
```

```python
import cv2

filename="3-Figure4-1.png" 
img_cv = cv2.imread(filename)
plt.figure(figsize=(10, 10))
plt.imshow(img_cv,cmap = 'gray')
plt.title('template'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:/Program Files/Tesseract-OCR/tesseract.exe'  # your path may be different
arabic_text=pytesseract.image_to_string(img_cv , lang='ara',config= ".")
print(arabic_text)
```

```python
tts = gTTS(arabic_text, lang='ar')
tts.save('arabic.mp3')
```
